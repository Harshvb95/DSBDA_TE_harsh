```python
#pie chart
colors=('red','yellow','blue','green')
a['cp'].unique()
```

```python
labels = '0','1','2','3'
sizes= a['cp'].value_counts()
```

```python
plt.hist(a['age'],color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
#pair plt
sns.set(style='ticks')
```

```python
sns.pairplot(a ,vars=['age','target','cp'])
```

```python
import matplotlib.pyplot as plt

# Extracting data
chol = a.iloc[:20, 4]  # Select all rows for the 'chol' column
age = a.iloc[:20,0]  # Select all rows for the 'age' column


# Plot
plt.figure(figsize=(10, 6))
plt.plot(age, chol, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Cholesterol Level')
plt.title('Cholesterol Level vs. Age')

# Show plot
plt.grid(True)
plt.show()
```

```python
#Scatter Plot
x = a['age']
y = a['chol']
plt.scatter(x,y,color='blue')
plt.xlabel('age')
plt.ylabel('cholestrol')
```

```python
#Heat Map
corr=a.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#box plot
import matplotlib.pyplot as plt

# Plot box plot for cholesterol level
plt.figure(figsize=(8, 6))  # Set the figure size
plt.boxplot(a['chol'], vert=False)    # Plot the box plot for cholesterol level horizontally
plt.title('Box Plot of Cholesterol Level')
plt.xlabel('Cholesterol Level')
plt.show()
```
